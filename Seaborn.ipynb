# Heatmaps
#### a heatamap is a graphical representation of data where the individual values contained in a matrix are represented as colors

```python
df=sns.load_dataset('flights')
```

```python
df.head()
```

```python
x=df.pivot_table(index='year',columns='month',values='passengers',aggfunc='sum')
```

```python
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x,cbar=False)
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='viridis')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='summer')
```

```python
sns.heatmap(x,cbar=False,linewidth=0.5,annot=True,fmt='d',cmap='winter')
```

# clustermap

```python
sns.clustermap(x,z_score=0,annot=True,row_cluster=False,metric='correlation')
```

# Distplot (distribution plot)
Displots are used to do univariate anslysis on a continuous/numerical data

Displot combines 3 things in one graph

1. Histogram
2. Kernel Density Estimation (PDF)
3. Rugplot

```python
titanic=sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.describe()
```

```python
titanic.isnull().sum()
```

```python
titanic['age'].fillna(titanic['age'].mean(),inplace=True)
```

```python
titanic.isnull().age.sum()
```

## Plotting Displot

```python
sns.displot(titanic['age'],kde=True)
```

```python
sns.displot(titanic['age'])
```

```python
sns.displot(titanic['age'],bins=100,kde=False)
```

# Plotting only the KDE

```python
sns.displot(titanic['age'],kind='kde')
```

# Plotting only Rugplot

```python
sns.distplot(titanic['age'],rug=True)
```

```python
sns.distplot(titanic['age'],hist=False,kde=False,rug=True)
```

## How to use the Distplot

```python
sns.distplot(titanic[titanic['survived']==1]['age'])
sns.distplot(titanic[titanic['survived']==0]['age'])
```
