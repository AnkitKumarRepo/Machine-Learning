# Pandas

= pandas refers to the pandas library,which is a popular open-source data manipulation and analysis tool.it provides data structure and function designed to make working with structured data,such as tabular data,easier,and more efficient
it has two primary data structure: the series and the DataFrame
pandas provide a wide range of functionalities for data manipulation and analysis ,including data cleaning,data filtering,aggregation ,merging,reshaping and more.it also integrate well with the other popular python library such as Numpy and Matplotlib
CSV(comma separated values) -plain text file that stores tabular data (rows and column)in a structured format.
