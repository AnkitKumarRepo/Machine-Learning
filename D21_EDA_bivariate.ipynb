# DistPlot 

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=True)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=True)
```

# HeatMap (categorical-categorical)

```python
titanic.head()
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
# percentane calculation for survival 
(titanic.groupby('Pclass').mean()['Survived'] *100)
```

```python
# percentane calculation for survival 
(titanic.groupby('Pclass').mean()['Survived'] *100).plot(kind='bar')
```

```python
#male femal survival
titanic.groupby('Sex').mean()['Survived']
```

```python
#plot
(titanic.groupby('Sex').mean()['Survived']).plot(kind='pie',autopct='%.2f')
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```
